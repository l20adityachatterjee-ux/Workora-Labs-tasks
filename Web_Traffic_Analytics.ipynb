```python
# ============================================================
#            WEB TRAFFIC ANALYTICS PROJECT
# ============================================================

# -----------------------------
# Import Libraries
# -----------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files

plt.style.use("ggplot")
sns.set_style("whitegrid")

# -----------------------------
# Upload Dataset
# -----------------------------
print("Upload your Web Traffic CSV file")

uploaded = files.upload()

df = pd.read_csv(next(iter(uploaded)))

# -----------------------------
# Basic Information
# -----------------------------
print("\n==============================")
print("DATASET PREVIEW")
print("==============================")

display(df.head())

print("\nDataset Shape")
print(df.shape)

print("\nDataset Info")
print(df.info())

print("\nMissing Values")
print(df.isnull().sum())

print("\nDuplicate Rows:", df.duplicated().sum())

# -----------------------------
# Data Cleaning
# -----------------------------

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Fill numerical missing values
num_cols = df.select_dtypes(include=['int64','float64']).columns

for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Fill categorical missing values
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    df[col].fillna("Unknown", inplace=True)

print("\nData Cleaning Completed Successfully!")

# -----------------------------
# Summary Statistics
# -----------------------------
print("\nSummary Statistics")

display(df.describe(include='all'))

# ============================================================
# VISUALIZATIONS
# ============================================================

# -----------------------------
# Page Views
# -----------------------------
if "PageViews" in df.columns:

    plt.figure(figsize=(8,5))

    sns.histplot(df["PageViews"], bins=20, kde=True)

    plt.title("Distribution of Page Views")

    plt.xlabel("Page Views")

    plt.ylabel("Frequency")

    plt.show()

# -----------------------------
# Session Duration
# -----------------------------
if "SessionDuration" in df.columns:

    plt.figure(figsize=(8,5))

    sns.histplot(df["SessionDuration"], bins=20, color="green", kde=True)

    plt.title("Session Duration Distribution")

    plt.xlabel("Session Duration (Minutes)")

    plt.ylabel("Frequency")

    plt.show()

# -----------------------------
# Traffic Source
# -----------------------------
if "TrafficSource" in df.columns:

    plt.figure(figsize=(9,5))

    sns.countplot(data=df,
                  x="TrafficSource",
                  order=df["TrafficSource"].value_counts().index)

    plt.xticks(rotation=30)

    plt.title("Traffic Sources")

    plt.show()

# -----------------------------
# Device Type
# -----------------------------
if "DeviceType" in df.columns:

    plt.figure(figsize=(7,5))

    sns.countplot(data=df,
                  x="DeviceType",
                  order=df["DeviceType"].value_counts().index)

    plt.title("Visitors by Device")

    plt.show()

# -----------------------------
# Browser
# -----------------------------
if "Browser" in df.columns:

    plt.figure(figsize=(8,5))

    sns.countplot(data=df,
                  x="Browser",
                  order=df["Browser"].value_counts().index)

    plt.xticks(rotation=20)

    plt.title("Browser Distribution")

    plt.show()

# -----------------------------
# Country
# -----------------------------
if "Country" in df.columns:

    plt.figure(figsize=(10,5))

    sns.countplot(data=df,
                  y="Country",
                  order=df["Country"].value_counts().index)

    plt.title("Visitors by Country")

    plt.show()

# -----------------------------
# Bounce Rate
# -----------------------------
if "Bounce" in df.columns:

    bounce = df["Bounce"].value_counts()

    plt.figure(figsize=(6,6))

    plt.pie(
        bounce,
        labels=bounce.index,
        autopct="%1.1f%%",
        startangle=90
    )

    plt.title("Bounce Rate")

    plt.show()

# -----------------------------
# Landing Pages
# -----------------------------
if "LandingPage" in df.columns:

    plt.figure(figsize=(10,5))

    sns.countplot(
        data=df,
        y="LandingPage",
        order=df["LandingPage"].value_counts().index
    )

    plt.title("Top Landing Pages")

    plt.show()

# -----------------------------
# Exit Pages
# -----------------------------
if "ExitPage" in df.columns:

    plt.figure(figsize=(10,5))

    sns.countplot(
        data=df,
        y="ExitPage",
        order=df["ExitPage"].value_counts().index
    )

    plt.title("Top Drop-off Pages")

    plt.show()

# -----------------------------
# Traffic Source vs Bounce
# -----------------------------
if "TrafficSource" in df.columns and "Bounce" in df.columns:

    plt.figure(figsize=(10,5))

    sns.countplot(
        data=df,
        x="TrafficSource",
        hue="Bounce"
    )

    plt.xticks(rotation=20)

    plt.title("Traffic Source vs Bounce")

    plt.show()

# -----------------------------
# Device vs Bounce
# -----------------------------
if "DeviceType" in df.columns and "Bounce" in df.columns:

    plt.figure(figsize=(8,5))

    sns.countplot(
        data=df,
        x="DeviceType",
        hue="Bounce"
    )

    plt.title("Device Type vs Bounce")

    plt.show()

# -----------------------------
# Session Duration vs Bounce
# -----------------------------
if "SessionDuration" in df.columns and "Bounce" in df.columns:

    plt.figure(figsize=(8,5))

    sns.boxplot(
        data=df,
        x="Bounce",
        y="SessionDuration"
    )

    plt.title("Session Duration vs Bounce")

    plt.show()

# -----------------------------
# Page Views vs Bounce
# -----------------------------
if "PageViews" in df.columns and "Bounce" in df.columns:

    plt.figure(figsize=(8,5))

    sns.boxplot(
        data=df,
        x="Bounce",
        y="PageViews"
    )

    plt.title("Page Views vs Bounce")

    plt.show()

# -----------------------------
# Correlation Heatmap
# -----------------------------
numeric = df.select_dtypes(include=np.number)

if not numeric.empty:

    plt.figure(figsize=(8,6))

    sns.heatmap(
        numeric.corr(),
        annot=True,
        cmap="coolwarm",
        fmt=".2f"
    )

    plt.title("Correlation Heatmap")

    plt.show()

# ============================================================
# USER JOURNEY ANALYSIS
# ============================================================

if "LandingPage" in df.columns and "ExitPage" in df.columns:

    print("\nTop User Journeys")

    journey = (
        df.groupby(["LandingPage","ExitPage"])
        .size()
        .reset_index(name="Users")
        .sort_values("Users",ascending=False)
    )

    display(journey.head(15))

# ============================================================
# KEY METRICS
# ============================================================

print("\n==============================")
print("KEY METRICS")
print("==============================")

print("Total Visitors :", len(df))

if "PageViews" in df.columns:
    print("Average Page Views :", round(df["PageViews"].mean(),2))

if "SessionDuration" in df.columns:
    print("Average Session Duration :", round(df["SessionDuration"].mean(),2),"minutes")

if "Bounce" in df.columns:
    bounce_rate = (df["Bounce"]=="Yes").mean()*100
    print("Bounce Rate :", round(bounce_rate,2),"%")

if "TrafficSource" in df.columns:
    print("\nTraffic Sources")
    print(df["TrafficSource"].value_counts())

if "DeviceType" in df.columns:
    print("\nDevice Usage")
    print(df["DeviceType"].value_counts())

if "LandingPage" in df.columns:
    print("\nTop Landing Pages")
    print(df["LandingPage"].value_counts().head())

if "ExitPage" in df.columns:
    print("\nTop Exit Pages")
    print(df["ExitPage"].value_counts().head())

# ============================================================
# BUSINESS INSIGHTS
# ============================================================

print("\n==============================")
print("BUSINESS INSIGHTS")
print("==============================")

print("1. Higher page views indicate stronger user engagement.")
print("2. Longer session durations generally reflect better content quality.")
print("3. High bounce rates suggest issues with landing page relevance or user experience.")
print("4. Traffic source analysis helps identify the most effective marketing channels.")
print("5. Device analysis can guide responsive website optimization.")
print("6. User journey mapping reveals the most common navigation paths.")
print("7. Frequently exited pages should be reviewed to reduce user drop-off.")
print("8. Improving pages with high exit rates can increase conversions and retention.")

print("\nProject Completed Successfully!")
```
